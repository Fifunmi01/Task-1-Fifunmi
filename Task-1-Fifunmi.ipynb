```python
# Project 1: Rule-Based AI Chatbot
# Powered by purely deterministic logic.

def start_chatbot():
    print("=== System Initialized: Rule-Based Chatbot Online ===")
    print("Welcome to my Interactive Portfolio.\n")

    # 1. The Knowledge Base (Numbered Keys)
    responses = {
        "1": "My name is Fifunmi, and this is my logic engine.",
        "2": "I am studying at Bowen University, Studying Computer Science",
        "3": "I am building a deterministic Rule-Based Chatbot for DecodeLabs.",
        "4": "I am operating at efficient level!"
    }

    # 2. The Infinite Loop
    while True:
        # Display the interactive menu
        print("Please select an option:")
        print("[1] Who are you?")
        print("[2] What is your education?")
        print("[3] What are you working on?")
        print("[4] System Status")
        print("[5] Exit")

        # Get the user's numbered choice
        choice = input("\nEnter a number (1-5): ").strip()

        # The Exit Strategy
        if choice == '5':
            print("\nBot: Shutting down the logic engine. Goodbye!")
            break

        # The Logic Engine & Fallback Mechanism
        reply = responses.get(choice, "Invalid selection. Please enter a number between 1 and 5.")

        # Output the response
        print(f"\nBot: {reply}\n")
        print("-" * 40) # Adds a clean visual divider between interactions

if __name__ == "__main__":
    start_chatbot()
```
